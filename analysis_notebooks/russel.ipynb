```python
deldf["batting_team"].unique()
```

```python
kkrdf = deldf[deldf['batting_team'] ==  'Kolkata Knight Riders']
```

```python
df = kkrdf[kkrdf['batsman'] == 'AD Russell']
```

```python
rusdf = df[df['match_id'] == 532 ]
```

```python
def phase(over):
    # over number in our dataset starts from 0
    if over <= 5:
        return 'Powerplay'
    elif over <= 14:
        return 'Middle'
    else:
        return 'Death'
```

```python
rusdf['phase'] = rusdf['over'].apply(lambda x: phase(x))
```

```python
rusdf = rusdf.filter(['over', 'ball', 'phase', 'batsman', 'batsman_runs'])
```
