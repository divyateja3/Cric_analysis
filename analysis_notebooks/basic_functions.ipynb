```python
import numpy as np
import pandas as pd
```

```python
Group_A = ['IND', 'AUS', 'BAN']
Group_B = ['ENG', 'NZ', 'PAK']
Group_C = ['RCB', 'CSK','MI','SRH','KKR']
Group_D = ['GL','LSG','DC','RR','PK']
```

```python
#basic_fixtures_b/w_2groups
for team_a in Group_A:
    for team_b in Group_B:

        print(team_a ,'VS', team_b)

for team_a in Group_A:
    for team_c in Group_C:
        print(team_a ,'VS', team_c)
```

```python
#IPL_FORMAT_fixtures
data = []
for i in Group_C:
    for j,k in zip(Group_C,Group_D):
        data.append((i,k))
        if i == j:
            continue
        data.append((i,j))
        
for l in Group_D:
    for m in Group_D:
        if l == m:
            continue
        data.append((l,m))

queue = Group_D.copy()
for i in Group_C:
    j = np.random.choice(queue)
    data.append((j,i))
    queue.remove(j)
print(len(data))       


fixtures = pd.DataFrame(data,columns=['home_team','away_team'])

fixtures['Match'] = fixtures.home_team + " VS " + fixtures.away_team
fixtures[(fixtures.home_team == "RCB")| (fixtures.away_team == "RCB")]
```
